```python
#Importing the matlplotlib for plotting the graphs

import matplotlib.pyplot as plt
```

```python
#Plot distribution of Solar Power

plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('Distibution of Generated Power (kW)')
plt.xlabel("Generated Power (kW)")
plt.ylabel('Frequency')
plt.show()
```

```python
#Plotting distribution for all other columns as well

columns = df.columns
for col in columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col],bins=30,kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Detecting outliers (if any)

plt.figure(figsize=(10,6))
sns.boxplot(x=df['generated_power_kw'])
plt.title("boxplot of Generated Power")
plt.show()
```

As we can see from the above boxplot, there were no outliers in the generated power.
